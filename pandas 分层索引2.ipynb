```python
colors=['红','黄','蓝']
species=['牛','马','猪']
mIndex=pd.MultiIndex.from_product([colors,species],names=['颜色','品种'])

frame2 = DataFrame(np.random.randn(3,9),columns=mIndex)
frame2.index.names=['序号']
frame2
```

```python
frame2.swaplevel(0,1,axis=1)
```

```python
frame2.swaplevel(0,1,axis=1).sortlevel(0,axis=1)
```
